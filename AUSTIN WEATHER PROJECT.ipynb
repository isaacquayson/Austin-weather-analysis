```python
# converting all object dtypes to float dtype and replacing errors with NaN
cols_to_convert = ['DewPointHighF',
       'DewPointAvgF', 'DewPointLowF', 'HumidityHighPercent',
       'HumidityAvgPercent', 'HumidityLowPercent',
       'SeaLevelPressureHighInches', 'SeaLevelPressureAvgInches',
       'SeaLevelPressureLowInches', 'VisibilityHighMiles',
       'VisibilityAvgMiles', 'VisibilityLowMiles', 'WindHighMPH', 'WindAvgMPH',
       'WindGustMPH', 'PrecipitationSumInches', 'Events']
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.dtypes #rechecking data types
```

```python
df.isnull().sum() # rechecking for null values 
```

```python
# calculating for percent NaN Iin each column and droping column(s) with 70% and above 
percent_nan = (df.isna().sum() / len(df)) * 100
print(round(percent_nan), 3)
```

```python
# since Events is showing 100% null values, it must be dropped
df.drop('Events', axis = 1, inplace = True)
```

```python
# since the rest of the columns show very low NaN percents,the NaN values are deleted
df.dropna(inplace = True)
```

# DATA VISUALISATIONS 


```python
# heatmap to see correlations 
plt.figure(figsize = (30, 16))
corr= df.corr()
matrix = np.triu(corr)
sns.heatmap(corr, annot = True, cmap = 'coolwarm', mask = matrix)
plt.show()
```

```python
# plotting temperatures over time
plt.suptitle('TEMPERATURES PLOT')
plt.figure(figsize = (20, 12))
plt.subplot(3,1,1)
sns.lineplot(x = df.Date,y = df.TempHighF, color = 'orange', label = 'TempHighF')
plt.subplot(3,1,2)
sns.lineplot(x = df.Date, y = df.TempAvgF, color = 'g', label = 'TempAvgF')
plt.subplot(3,1,3)
sns.lineplot(x =df.Date, y=df.TempLowF, color = 'b', label = 'TempLowF')
plt.show()
```

```python
# boxplot to see the outliers in the high, avg and low temperatures and the distributions 
df.boxplot(column=['TempHighF', 'TempAvgF', 'TempLowF'], color='green', showfliers = True)
```

```python
# boxplot to see the outliers in the high, avg and low humidty and the distributions 
df.boxplot(column=['HumidityHighPercent','HumidityAvgPercent', 'HumidityLowPercent',], color='red', showfliers = True)
```

```python
# plotting humidities over time
plt.suptitle('TEMPERATURES PLOT')
plt.figure(figsize = (20, 12))
plt.subplot(3,1,1)
sns.lineplot(x = df.Date,y = df.HumidityHighPercent, color = 'orange', label = 'HumidityHighPercent')
plt.subplot(3,1,2)
sns.lineplot(x = df.Date, y = df.HumidityAvgPercent, color = 'g', label = 'HumidityAvgPercent')
plt.subplot(3,1,3)
sns.lineplot(x =df.Date, y=df.HumidityLowPercent, color = 'b', label = 'HumidityLowPercent')
plt.show()
```

```python
# scatter plot between average temperature and average humidity to see if negative or positive correlation
sns.regplot(x = df.TempAvgF, y = df.HumidityAvgPercent, color = 'green')
plt.show()
```

```python
# scatter plot between higher temperature and higher humidity to see if negative or positive correlation
sns.regplot(x = df.TempHighF, y = df.HumidityHighPercent, color = 'crimson')
plt.show()
```

```python
# plotting sea level pressures over time
plt.suptitle('Sea level pressures PLOT')
plt.figure(figsize = (20, 12))
plt.subplot(3,1,1)
sns.lineplot(x = df.Date,y = df.SeaLevelPressureHighInches, color = 'orange', label = 'SeaLevelPressureHighInches')
plt.subplot(3,1,2)
sns.lineplot(x = df.Date, y = df.SeaLevelPressureAvgInches, color = 'g', label = 'SeaLevelPressureAvgInches')
plt.subplot(3,1,3)
sns.lineplot(x =df.Date, y=df.SeaLevelPressureLowInches, color = 'b', label = 'SeaLevelPressureLowInches')
plt.show()
```

```python
# boxplot to see the outliers in the high, avg and low visibility and the distributions 
df.boxplot(column=['VisibilityHighMiles','VisibilityAvgMiles', 'VisibilityLowMiles',], color='blue', showfliers = True)
```

```python
# scatter plot between average humidity and average visibility to see if negative or positive correlation
sns.regplot(x = df.VisibilityAvgMiles, y = df.HumidityAvgPercent)
plt.show()
```

```python
# boxplot to see the outliers in the high, avg and low wind and the distributions 
df.boxplot(column=['WindHighMPH', 'WindAvgMPH','WindGustMPH'], color='green', showfliers = True)
plt.show()
```
